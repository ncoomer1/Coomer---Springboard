# Decision Tree

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

```

```python
maternal_health=pd.read_csv(r"C:\Users\nicol\Desktop\Capstone 2 - Data Wrangling\Maternal Health Risk Data Set.csv")
```

```python
print(maternal_health)
```

```python
features = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']

X = maternal_health[features]
y = maternal_health['RiskLevel']
```

```python
X.shape , y.shape
```

```python
X.isnull().sum()
```

```python
X.info()
```

```python
# split the dataset

from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2 , stratify=y)
```

```python
X_train.shape , y_train.shape , X_test.shape, y_test.shape
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
```

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtree,
                   feature_names=X.columns,
                   class_names=['low risk', 'mid risk' , 'high risk'],
                   filled=True)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
#function to print classification report and get confusion matrix in a proper format

def metrics_score(actual, predicted):
    print(classification_report(actual, predicted))
    cm = confusion_matrix(actual, predicted)
    plt.figure(figsize=(8,5))
    sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels=['low risk', 'mid risk' , 'high risk'], yticklabels=['low risk', 'mid risk' , 'high risk'])
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
# Checking performance on the training data
y_pred_train = dtree.predict(X_train)
metrics_score(y_train,y_pred_train)
```

```python
# Checking performance on the test data
y_pred_test = dtree.predict(X_test)
metrics_score(y_test,y_pred_test)
```

```python
# Create the parameter grid based on the results of random search 
params = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ["gini", "entropy"]
}
```

```python
#Grid Search for the Decision Tree
param_dtc = {'min_samples_split': np.arange(2, 10), 'min_samples_leaf': 
np.arange(.05, .2), 'max_leaf_nodes': np.arange(2, 30) , 'max_depth' : np.arange(2,30)}
```

```python
# Instantiate the grid search model
grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), 
                           param_grid=param_dtc, 
                           cv=5, n_jobs=-1, verbose=1, scoring = "accuracy" )
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
dtree_tuned = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=12, min_samples_leaf=0.05).fit(X_train , y_train)
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtree_tuned,
                   feature_names=X.columns,
                   class_names=['low risk', 'mid risk' , 'high risk'],
                   filled=True)
```

```python
y_pred_train = dtree_tuned.predict(X_train)
metrics_score(y_train,y_pred_train)
```

```python
y_pred_test = dtree_tuned.predict(X_test)
metrics_score(y_test,y_pred_test)
```
